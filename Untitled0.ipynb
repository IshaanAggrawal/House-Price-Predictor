```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
from xgboost import XGBRegressor
```

```python
dataset = pd.read_csv('BostonHousing.csv', sep=',')
dataset
```

```python
dataset.isnull()
```

```python
print(dataset.dtypes)
corelation=dataset.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(corelation,annot=True,cbar=True,square=True,fmt='.1f',annot_kws={'size':8},cmap='Blues')
plt.show()
```

```python
dataset = dataset.dropna()
x=dataset.drop('medv',axis=1)
y=dataset['medv']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model_prediction=model.predict(x_test)
model_prediction
```

```python
score1=r2_score(model_prediction,y_test)
score1
```

```python
score2=mean_squared_error(model_prediction,y_test)
score2
```

```python
model2=XGBRegressor()
model2.fit(x_train,y_train)
```

```python
model_prediction2=model2.predict(x_test)
model_prediction2
```

```python
score1=r2_score(model_prediction2,y_test)
score1
```

```python
score2=mean_squared_error(model_prediction2,y_test)
score2
```
